# Remark 4.10

```python
prof410 = Profile([[0, 1, 2, 3], [1, 2, 3, 0], [0, 1, 2, 3], [0, 1, 3, 2]], rcounts = [1, 1, 1, 1], cmap = {0:"a", 1:"b", 2:"c", 3:"d"})
prof410.display()
prof410.display_margin_graph()  
mg = prof410.margin_graph()
print("Margin graph with Split cycle defeat relation:")
mg.display_with_defeat(split_cycle_defeat(mg))
```

```python
prof410prime = Profile([[0, 1, 2, 3], [1, 2, 3, 0], [2, 3, 0, 1], [3, 0, 1, 2]], rcounts = [1, 1, 1, 1], cmap = {0:"a", 1:"b", 2:"c", 3:"d"})
prof410prime.display()
prof410prime.display_margin_graph()  
mg = prof410prime.margin_graph()
print("Margin graph with Split Cycle defeat relation:")
mg.display_with_defeat(split_cycle_defeat(mg))
```
